```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("F:\\Machine Learning\\linear regression\\train_cleaned.csv")
```

```python
data.head()
```

```python
x = data.drop('Item_Outlet_Sales',axis=1)
y=data['Item_Outlet_Sales']
```

```python
x.shape , y.shape
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_absolute_error as mae 
```

```python
from sklearn.model_selection  import train_test_split 
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=56)
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
train_pred = lr.predict(x_train)
train_mae = mae(train_pred,y_train)
train_mae
```

```python
test_pred = lr.predict(x_test)
test_mae = mae(test_pred,y_test)
test_mae
```

```python
lr.coef_
```

```python
plt.figure(figsize=(8,6),dpi=120,facecolor='w',edgecolor='b')
x1 = range(len(x.columns))
y= lr.coef_
plt.bar(x1,y)
plt.title('Co-efficient plot')
plt.xlabel('Variables')
plt.xlabel('Co-efficirnts')
plt.show()
```

```python
residuals = pd.DataFrame({
    'fitted values' : y_test,
    'predicted values' : test_pred,
})

residuals['residuals'] = residuals['fitted values'] - residuals['predicted values']
residuals.head()
```

```python
residuals.shape
```

```python
plt.figure(figsize=(10,5),dpi=120,facecolor= 'w',edgecolor='b')
f = range(0,2131)
k = [0 for i in range(0,2131)]
plt.scatter(f,residuals.residuals[:],label = 'Residuals')
plt.plot(f,k,color='r',label='Regression Line')
plt.title('Residual plot')
plt.ylim(-4000,4000)
plt.show()
```

```python
# Histogram for distribution
plt.figure(figsize=(10, 6), dpi=120, facecolor='w', edgecolor='b')
plt.hist(residuals.residuals, bins = 150)
plt.xlabel('Error')
plt.ylabel('Frequency')
plt.title('Distribution of Error Terms')
plt.show()
```

```python
from statsmodels.graphics.gofplots import qqplot
fig, ax = plt.subplots(figsize = (10,8),dpi=120,facecolor ='w',edgecolor='r')
qqplot(residuals.residuals,line='s',ax = ax)
plt.ylabel('Residual Quantiles')
plt.xlabel('Ideal Scaled Quantiles')
plt.title('Checking distribution of Residual Errors')
plt.show()
```
