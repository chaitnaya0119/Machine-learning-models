```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate random height data with a normal distribution
mean_height = 170  # Mean height in centimeters
std_deviation = 10  # Standard deviation in centimeters
num_samples = 1000  # Number of data points

height_data = np.random.normal(mean_height, std_deviation, num_samples)

# Create a histogram to visualize the distribution
plt.figure(figsize=(8, 6))
plt.hist(height_data, bins=20, density=True, alpha=0.6, color='blue')

# Add a normal distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean_height, std_deviation)
plt.plot(x, p, 'k', linewidth=2)

plt.title('Normal Distribution of Heights')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.legend(['Normal Distribution', 'Height Data'])

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data for two variables
np.random.seed(42)
x = np.random.rand(50) * 10  # Variable 1 (x-axis)
y = 2 * x + np.random.randn(50) * 2  # Variable 2 (y-axis) with some noise

# Create a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Data Points')
plt.title('Bivariate Analysis: Scatter Plot')
plt.xlabel('Variable 1 (X)')
plt.ylabel('Variable 2 (Y)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Generate synthetic data for three variables
np.random.seed(42)
num_samples = 100
variable1 = np.random.randn(num_samples) * 2
variable2 = variable1 + np.random.randn(num_samples)
variable3 = np.random.randn(num_samples) * 3

# Create a DataFrame from the variables
data = {
    'Variable 1': variable1,
    'Variable 2': variable2,
    'Variable 3': variable3
}
df = pd.DataFrame(data)

# Create a pair plot (scatter plot matrix)
sns.pairplot(df)
plt.suptitle('Multivariate Analysis: Pair Plot', y=1.02)
plt.show()
```

```python
import numpy as np

# Example data: Three variables with five observations each
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 4, 5, 6])
z = np.array([5, 4, 3, 2, 1])

# Stack variables together to create a data matrix
data_matrix = np.vstack((x, y, z))

# Calculate the covariance matrix
covariance_matrix = np.cov(data_matrix)

print("Covariance Matrix:")
print(covariance_matrix)
```
